```python
import yfinance as yf
import matplotlib.pyplot as plt

def plot_btc():
    #  Bitcoin  yfinance
    df_btc = yf.Ticker('BTC-USD').history(period='max')

    #  Bitcoin
    plt.figure(figsize=(12, 6))
    plt.plot(df_btc.index, df_btc['Close'], label='BTC-USD Closing Price')
    plt.title('BTC-USD Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.show()

#  Bitcoin
plot_btc()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

def plot_btc_with_splits():
    #  Bitcoin
    df_btc = yf.Ticker('BTC-USD').history(period='max')
    
    #  (chunk size)  8 
    num_rows = df_btc.shape[0]
    chunk_size = num_rows // 8
    
    # 
    split_points = np.arange(0, num_rows, chunk_size)
    
    # 
    plt.figure(figsize=(12, 6))
    plt.plot(df_btc.index, df_btc['Close'], label=' BTC-USD')
    
    # 
    for point in split_points:
        plt.axvline(df_btc.index[point], color='red', linestyle='--', alpha=0.7, label='' if point == 0 else "")
    
    # 
    plt.title(' BTC-USD  8 ')
    plt.xlabel('')
    plt.ylabel(' (USD)')
    plt.legend()
    
    # 
    plt.show()

# 
plot_btc_with_splits()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

def plot_btc_with_splits():
    #  Bitcoin
    df_btc = yf.Ticker('BTC-USD').history(period='max')
    
    #  (chunk size)  8 
    num_rows = df_btc.shape[0]
    chunk_size = num_rows // 8
    
    # 
    split_points = np.arange(0, num_rows, chunk_size)
    
    # 
    plt.figure(figsize=(12, 6))
    plt.plot(df_btc.index, df_btc['Close'], label=' BTC-USD')
    
    # 
    for point in split_points:
        plt.axvline(df_btc.index[point], color='red', linestyle='--', alpha=0.7)
        # 
        plt.text(df_btc.index[point], df_btc['Close'][point], df_btc.index[point].strftime('%d-%m-%Y'), rotation=90, verticalalignment='center', horizontalalignment='right', fontsize=8, color='red')
    
    # 
    plt.title(' BTC-USD  8 ')
    plt.xlabel('')
    plt.ylabel(' (USD)')
    plt.legend()
    
    # 
    plt.show()

# 
plot_btc_with_splits()
```

```python
import yfinance as yf
import numpy as np

def get_btc_split_dates():
    #  Bitcoin
    df_btc = yf.Ticker('BTC-USD').history(period='max')
    
    #  (chunk size)  8 
    num_rows = df_btc.shape[0]
    chunk_size = num_rows // 8
    
    # 
    split_points = np.arange(0, num_rows, chunk_size)
    
    # 
    split_dates = [df_btc.index[point].strftime('%d-%m-%Y') for point in split_points]
    
    # 
    return split_dates

#  8 
dates = get_btc_split_dates()
print(" 8 :")
for date in dates:
    print(date)
```
