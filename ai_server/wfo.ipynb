```python
import yfinance as yf
import matplotlib.pyplot as plt

def plot_btc():
    # โหลดข้อมูลราคาของ Bitcoin โดยใช้ yfinance
    df_btc = yf.Ticker('BTC-USD').history(period='max')

    # สร้างกราฟของราคาของ Bitcoin
    plt.figure(figsize=(12, 6))
    plt.plot(df_btc.index, df_btc['Close'], label='BTC-USD Closing Price')
    plt.title('BTC-USD Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.show()

# เรียกใช้ฟังก์ชันเพื่อแสดงกราฟของราคาของ Bitcoin
plot_btc()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

def plot_btc_with_splits():
    # ดาวน์โหลดข้อมูลประวัติของ Bitcoin
    df_btc = yf.Ticker('BTC-USD').history(period='max')
    
    # คำนวณจำนวนแถวทั้งหมดและขนาดของแต่ละช่วงที่จะแบ่ง (chunk size) สำหรับ 8 ช่วง
    num_rows = df_btc.shape[0]
    chunk_size = num_rows // 8
    
    # กำหนดจุดแบ่ง
    split_points = np.arange(0, num_rows, chunk_size)
    
    # สร้างกราฟ
    plt.figure(figsize=(12, 6))
    plt.plot(df_btc.index, df_btc['Close'], label='ราคาปิด BTC-USD')
    
    # ทำเครื่องหมายจุดแบ่งบนกราฟ
    for point in split_points:
        plt.axvline(df_btc.index[point], color='red', linestyle='--', alpha=0.7, label='จุดแบ่ง' if point == 0 else "")
    
    # เพิ่มชื่อเรื่องและฉลาก
    plt.title('ราคาของ BTC-USD ตลอดเวลาพร้อม 8 ช่วง')
    plt.xlabel('วันที่')
    plt.ylabel('ราคา (USD)')
    plt.legend()
    
    # แสดงกราฟ
    plt.show()

# เรียกใช้ฟังก์ชันเพื่อแสดงกราฟ
plot_btc_with_splits()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

def plot_btc_with_splits():
    # ดาวน์โหลดข้อมูลประวัติของ Bitcoin
    df_btc = yf.Ticker('BTC-USD').history(period='max')
    
    # คำนวณจำนวนแถวทั้งหมดและขนาดของแต่ละช่วงที่จะแบ่ง (chunk size) สำหรับ 8 ช่วง
    num_rows = df_btc.shape[0]
    chunk_size = num_rows // 8
    
    # กำหนดจุดแบ่ง
    split_points = np.arange(0, num_rows, chunk_size)
    
    # สร้างกราฟ
    plt.figure(figsize=(12, 6))
    plt.plot(df_btc.index, df_btc['Close'], label='ราคาปิด BTC-USD')
    
    # ทำเครื่องหมายจุดแบ่งบนกราฟและแสดงวันเดือนปีของแต่ละจุด
    for point in split_points:
        plt.axvline(df_btc.index[point], color='red', linestyle='--', alpha=0.7)
        # แสดงวันเดือนปีที่แต่ละจุดแบ่ง
        plt.text(df_btc.index[point], df_btc['Close'][point], df_btc.index[point].strftime('%d-%m-%Y'), rotation=90, verticalalignment='center', horizontalalignment='right', fontsize=8, color='red')
    
    # เพิ่มชื่อเรื่องและฉลาก
    plt.title('ราคาของ BTC-USD ตลอดเวลาพร้อม 8 ช่วง')
    plt.xlabel('วันที่')
    plt.ylabel('ราคา (USD)')
    plt.legend()
    
    # แสดงกราฟ
    plt.show()

# เรียกใช้ฟังก์ชันเพื่อแสดงกราฟ
plot_btc_with_splits()
```

```python
import yfinance as yf
import numpy as np

def get_btc_split_dates():
    # ดาวน์โหลดข้อมูลประวัติของ Bitcoin
    df_btc = yf.Ticker('BTC-USD').history(period='max')
    
    # คำนวณจำนวนแถวทั้งหมดและขนาดของแต่ละช่วงที่จะแบ่ง (chunk size) สำหรับ 8 ช่วง
    num_rows = df_btc.shape[0]
    chunk_size = num_rows // 8
    
    # กำหนดจุดแบ่ง
    split_points = np.arange(0, num_rows, chunk_size)
    
    # รวบรวมวันเดือนปีที่ตรงกับแต่ละจุดแบ่ง
    split_dates = [df_btc.index[point].strftime('%d-%m-%Y') for point in split_points]
    
    # แสดงวันเดือนปีของแต่ละจุดแบ่ง
    return split_dates

# เรียกใช้ฟังก์ชันเพื่อรับวันเดือนปีที่แบ่งออกเป็น 8 ช่วง
dates = get_btc_split_dates()
print("วันเดือนปีที่แบ่งออกเป็น 8 ช่วง:")
for date in dates:
    print(date)
```
